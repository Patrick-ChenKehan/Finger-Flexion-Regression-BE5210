```python
raw_sub_1 = scipy.io.loadmat('./datasets/sub1_comp.mat')
raw_sub_1['train_data'].shape
raw_sub_1
```

```python
files = ['datasets/sub1_comp.mat', 'datasets/sub2_comp.mat', 'datasets/sub3_comp.mat']

for i in range(3):
    file = files[i]
    raw = scipy.io.loadmat(file)
    data = raw['train_data'] 
    labels = raw_sub_1['train_dg'].argmax(axis=1)
    ch_names = ['ch' + str(i) for i in range(1, data.shape[1] + 1)] + ['STI101']
    
    idx_change = np.array([labels[0]] + list(labels)) != np.array(list(labels) + [labels[-1]])
    index = np.insert(np.where(idx_change == True), 0, 1)
    stim = np.zeros(len(labels))
    stim[index] = labels[index] + 1
    data_fif = np.concatenate([data * 1e-6, stim.reshape(-1, 1)], axis=1)
    
    ch_types = [*['eeg' for _ in range(data.shape[1])], 'stim']

    sampling_freq = 1000  # in Hertz
    
    info = mne.create_info(ch_names=ch_names, ch_types=ch_types, sfreq=sampling_freq)

    User_raw = mne.io.RawArray(data_fif.T, info)
    User_raw.save(f'datasets/S{i + 1}R01.raw.fif', overwrite=True)
```
